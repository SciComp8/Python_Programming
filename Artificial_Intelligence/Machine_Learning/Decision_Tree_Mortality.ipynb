# Train The Decision Tree Classifier

## Examine The Case-Control Distribution

```python
# read data, and describe the distribution per variable
import pandas as pd
all_addressed_data = pd.read_csv("/Users/xyz/Documents/data/all_addressed_data.csv",
                                 index_col=False, header=0)

# how many cases and controls
n_case = len(all_addressed_data[all_addressed_data['label']==1])
n_control = len(all_addressed_data[all_addressed_data['label']==0])

print("case",n_case)
print("control",n_control)
```

```python
# Make barplts to show case-control distributions
import matplotlib.pyplot as plt

plt.bar(["Cases (death)","Controls (alive, discharge)"],[n_case,n_control])  
plt.ylabel("number of subjects")
plt.show
```

## Examine The Missing Distribtuion

```python
missing_dis = all_addressed_data.isnull().sum()/len(all_addressed_data)
print(missing_dis)
```

```python
# drop feature "Capillary refill rate" due to its high missingness
all_addressed_data = all_addressed_data.drop(columns=["Capillary refill rate"])

# examine the feature distribution: count, mean, min, max, std, ...
feature_distribution = all_addressed_data.drop(columns = ['label']).describe()
print(feature_distribution)
```

## Split Data Into 90% Training And 10% Testing Sets

```python
from sklearn.model_selection import train_test_split 
y = all_addressed_data["label"].values
X = all_addressed_data.drop(columns=["Subj","label"]).values

feature_used = all_addressed_data.drop(columns=["Subj","label"]).columns.tolist()
print("feature_used",feature_used)

# random_state to keep the spliting is reproducible
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=0) 
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# impute the missing values
from sklearn.impute import SimpleImputer
import numpy as np
imputer = SimpleImputer(missing_values = np.nan, strategy = "mean")
imputer.fit(X_train)

X_train = np.array(imputer.transform(X_train),dtype = np.float32)
X_test = np.array(imputer.transform(X_test),dtype = np.float32)
```

## Train The Decision Tree Classifier Using Training Data

```python
from sklearn.tree import DecisionTreeClassifier

# definine a tree
# sklearn DecisionTreeClassifier-----describe clearly--parameters, attributes
# https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html 
tree = DecisionTreeClassifier(criterion = 'gini', class_weight = 'balanced', random_state = 0)

# train the tree
tree.fit(X_train, y_train)

# use the trained classifier to make predictions in the test set
y_pred = tree.predict(X_test)
print(y_pred)

# evaluate the prediciton performance
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))

# support: the number of true instances for each label
# macro avg: calculate metrics for each label and find their unweighted mean. This does not take label imbalance into account
# weighted avg: calculate metrics for each label, and find their weighted mean.
```

```python
from sklearn.metrics import roc_auc_score,roc_curve

y_proba = tree.predict_proba(X_test)[:,1] 

# create true and false positive rates
fpr_dt, tpr_dt, threshold_dt = roc_curve(y_test,y_proba)
```

```python
# plot ROC Curve
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6)) 
plt.title('Revceiver Operating Characterstic') 
plt.plot(fpr_dt, tpr_dt) 
plt.plot([0,1],ls='--') 
plt.plot([0,0],[1,0],c='.5') 
plt.plot([1,1],c='.5')
plt.xlabel('False Positive Rate') 
plt.ylabel('True Positive Rate') 
plt.show()
```

```python
# calculate area under the curve
auc_lr = roc_auc_score(y_test, y_proba) 
print(auc_lr)
```

# Tune Model Parameters Using Grid Search

Goal: search for the optimal model parameters.

We tune the following parameter:
(1) max_depth: the maximum depth of the tree
(2) min_samples_split
(3) min_samples_leaf
(4) max_leaf_nodes
...

Reference: https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html#sklearn.tree.DecisionTreeClassifier.get_depth 


## Train the Model Using Grid Search With Cross-Validation

```python
# the goal is to identify the best hyper-parameters of the model
from sklearn.model_selection import GridSearchCV

tree = DecisionTreeClassifier(criterion = 'gini', class_weight = 'balanced', random_state = 0)

grid = GridSearchCV(tree, 
                    param_grid={"max_depth":[2,3,4,5,6,7,8,9,10]},
                    scoring = "accuracy",
                    cv = 5,
                    return_train_score = True
                   )
grid.fit(X_train, y_train)

# obtain grid results
scores = pd.DataFrame(grid.cv_results_)
```

```python
scores
```

```python
# visulization of grid search
scores.plot(x = "param_max_depth",y = "mean_train_score",yerr = "std_train_score")
scores.plot(x = "param_max_depth",y = "mean_test_score",yerr = "std_test_score",ax = plt.gca()) 
# ax = plt.gca() keep plot at the samp axis
plt.legend()
plt.show()
```

```python
grid.best_estimator_ # obtain the best classifier

# use the best model to predict
y_pred_grid = grid.best_estimator_.predict(X_test)

print(classification_report(y_test, y_pred_grid))
```
